#**Data Product Development**


Selain membuat analisis, data analyst/science dapat membuat product yaitu sebuah dashboard.

image.png

Dashboard bisa mempercepat alur komunikasi satu sama lain. Selain itu, dashboard bisa mengeliminasi kebutuhan seseorang yang melakukan pengecekan data terus.

#**Lookers**


Untuk pembuatan dashboard dapat menggunakan lookers.

image.png

Dari data lookers diatas dapat dilihat hubungannya / mana yang semakin unggul yaitu semakin atas (semakin baik (mudah digunakan)). Jika sebelah kanan (semakin visioner). Dua poin diatas masih mendominasi keseluruhan data.

**Lookers gratis, mudah digunakan, serta populer dikalangan Tech.**

#**METRIC VS DIMENSION**

Case : coba hitung berapa banyak jumlah nasabah di tiap kabupaten.

Banyak nasabah : metric
di tiap kabupaten : dimension

kita menghitung metric berdasarkan dimension

metric : bisa dihitung/aggregasi
