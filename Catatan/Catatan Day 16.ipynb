#**AGGREGATION**


SQL Aggregation merupakan sebuah tugas untuk mengkoleksi set value dengan mengembalikan single value. Hal ini selesai dengan membantu pemfungsian aggregate seperti SUM, COUNT, dan AVG. Sebagai contoh, di database products, kamu mungkin ingin mengkalkulasi rata-rata harga di seluruh inventaris.

image.png

Pada aggregaton terdapat GROUP BY dan HAVING CLAUSES

count()
sum()
AVG()
MIN()
MAX()

##COUNT

COUNT(*)

menghitung semua rows pada table

COUNT(DISTINC)

menghitung distinc rows di spesifik kolom tertentu

COUNT with GROUP BY

Group by digunakan untuk aggregate value berdasarkan kategori penting/beberapa kategori.

HAVING

kondisi seperti where (kondisi) yang sudah dibentuk group by
