#**RDBMS**

RDBMS (Relational Database Management System)

- Database = koleksi data untuk mengorganisasi secara sistematik sehingga dapat diakses lebih mudah dan data mudah terupdate.
- DBMS = Software yang di desain untuk membuat database dan menampilkan operasi dari database.
- RDBMS = DBMS yang data storenya merupakan tabular form dan relasi antara table

## Terminologies di RDBMS

- Table = data yang tersimpan dari bentuk row dan columns
- Record/Row = setiap entry data dalam tabel
- Column = informasi spesifik yang memiliki record dalam tabel
- Query = sebuah request mengambil beberapa data spesifik di database
- SQL = Stuctured Query Language

##Primary Key

Sebuah kolom atau beberapa set kolom yang memiliki value dengan unique di setiap row dalam tabel.

Aturannya : 
- Seiap tabel hanya boleh memiliki satu primary key
- Dalam satu tabel tidak boleh terdiri dari 3 atau 4 primary key
- Gaboleh terdiri dari null value
- Harus terdiri dari uniques value

- Intuitif

## Komponen SQL
Terdiri dari 3 komponen yang diperlukan untuk manage, maintain, dan digunakan dalam database
- Data definition Language -> Ditunjukan untuk mengupdate tabel, mendefine tabel, membuat tabel, drop tabel, nambahin kolom, dsb.
- Data Manipulation Language -> manipulasi datanya seperti masukin data ke dalam tabel, delete rownya, dll.
- Data Control Language ->  satu diantara kelompok perintah yang digunakan untuk melakukan kontrol terhadap privilage data.

**SQL Environment**
