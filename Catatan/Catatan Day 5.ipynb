**CATEGORICAL DATA ENCODING**

Merubah data kategorikal menjadi numerik.

```python
#Label encoding (2 kategori(1 dan 0 saja))
df['sex'] = df['sex'].astype('category').cat.codes
df.head(3)
```

```python
#One Hot Encoding
#Memberikan nilai unik setiap value
dummies_embark_town = pd.get_dummies(df['embark_town'],prefix='embark_town')
dummies_embark_town.head()
```

```python
#Cocok digunakan jika valuenya dikit
#Semakin banyak value akan membentuk kolom baru sesuai jumlah kolom
#Jika terlalu banyak akan terjadi sparse data
df = pd.concat([df, dummies_embark_town], axis=1)
df.head()
```

```python
#Case
#Jika menggunakan one hot
#Kota pulan jawa -> 50 kota -> 50 kolom baru
#Cara menyelesaikannya menggunakan grouping
#Contoh jadiin provinsi sehingga jadi 6 value
#Contoh groupin jadi kota, kab, ibu kota
```

```python
#Frequency Encoding
#Mengisi sesuai beberapa kali value muncul
freq_et = df['embark_town'].value_counts().reset_index()
freq_et.rename(columns={"index": "embark_town", "embark_town": "freq_embark_town"}, inplace = True)
freq_et['pct_embark_town'] = round((freq_et['freq_embark_town']/freq_et['freq_embark_town'].sum())*100,2)
freq_et
```

```python
df = df.merge(freq_et[['embark_town','pct_embark_town']], on='embark_town', how='inner')
df.head()
```

OUTLIER Z-SCORE
image.png

setelah 3 outlier max dan setelah -3 outlier min. Konsep Z core itu membuang row

```python
df = sns.load_dataset('titanic')
df.head(3)
```

```python
#METODE z-SCORE
#Import stats terlebih dahulu
from scipy import stats
```

```python
#BOXPLOT ORIGINAL
sns.boxplot(df['fare'],color='green',orient='h')
```

```python
df['fare zscore'] = abs(stats.zscore(df['fare']))
```

```python
df.head(3)
```

```python
print('Jumlah baris sebelum handling outlier pada kolom Fare:', len(df))
df['fare zscore'] = abs(stats.zscore(df['fare']))
df = df[df['fare zscore'] < 3]
print(f'Jumlah baris setelah memfilter outlier:', len(df))
```

```python
#BOX PLOT SETELAH z SCORE
#Masih ada outlier tapi sedikit nilainya
sns.boxplot(df['fare'],color='green',orient='h')
```

```python
#Handing outlier dengan IQR
df = sns.load_dataset('titanic')
df.head(3)
```

```python
print('Jumlah baris sebelum handling outlier pada kolom Fare:', len(df))
Q1 = df['fare'].quantile(0.25)
Q3 = df['fare'].quantile(0.75)
IQR = Q3 - Q1
low_limit = Q1 - (1.5 * IQR)
high_limit = Q3 + (1.5 * IQR)
filtered_entries = ((df['fare'] >= low_limit) & (df['fare'] <= high_limit))
df = df[filtered_entries]
print(f'Jumlah baris setelah memfilter outlier:', len(df))
df.head()
```
