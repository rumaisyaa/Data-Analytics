```python
numbers=[4,5,10,20,40,60,80]

# ambil element/value >= 10, then masukan ke dalam new_list
new_list = []
for i in numbers:
  if i>=10:
    new_list.append(i)
    print(new_list)

print('Akhir : ',new_list)
```

```python
def hitung(a, b, status):
  if status == 'tambah':
    return a+b
  elif status == 'kali':
    return a*b
  elif status == 'bagi':
    return a/b
```

```python
hitung(6, 9, 'tambah')
```

```python
hitung(50,80, 'bagi')
```

```python
hitung(7,5, 'kali')
```
