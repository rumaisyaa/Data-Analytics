### Task 1: Compute mean values and variance for all values

```python
#Menentukan mean dan variance value AGE
#Mencari mean value dengan code (.mean()) dan variance dengan code (.var())
mean = df['AGE'].mean()
var = df['AGE'].var()

#Print hasil mean dan variance sehingga diketahui nilainya
print(f"Mean = {mean}\nVariance = {var}")
```

Varians adalah ukuran dari seberapa jauh penyebaran data dari nilai rata-ratanya.Semakin besar nilai varians, semakin jauh data yang kita gunakan tersebar dari nilai rata-ratanya. Dari value 'AGE' terlihat bahwa nilai variansnya lebih besar dari rata-ratanya sehingga data yang digunakan jauh tersebar dari nilai rata-ratanya.

### Task 2: Plot boxplots for BMI, BP and Y depending on gender

```python
#Menentukan bloxplot BMI berdasarkan gender atau value SEX
df.boxplot(column='BMI',by='SEX')
#Membuat bloxplot secara vertical
plt.xticks(rotation='vertical')
#Boxplot ditampilkan
plt.show()
```

Median dari data berat badan berdasarkan gender male (1) dan female (2) terlihat hampir sejajar atau hampir saling beririsan. Namun, jika ditinjau lebih lanjut, nilai median gender female sedikit lebih tinggi dari gender male. Kemudian, variansi berat badan pada gender female lebih banyak dibandingkan gender male.

```python
#Menentukan bloxplot BP berdasarkan gender atau value SEX
df.boxplot(column='BP',by='SEX')
#Membuat bloxplot secara vertical
plt.xticks(rotation='vertical')
#Boxplot ditampilkan
plt.show()
```

Nilai tengah (median) pada value BP atau tekanan darah tidak sejajar atau tidak beririsan. Kemudian varians data tekanan darah pada gender pria (male) lebih besar/banyak dari gender female.

```python
#Menentukan bloxplot Y berdasarkan gender atau value SEX
df.boxplot(column='Y',by='SEX')
#Membuat bloxplot secara vertical
plt.xticks(rotation='vertical')
#Boxplot ditampilkan
plt.show()
```

Median pada value 'Y' atau ukuran kualitatif perkembangan penyakit selama satu tahun hampir beririsan antara male dan female.Kemudian, varians male dalam value 'Y' lebih banyak dibanding female. Dengan demikian, nilai maksimum pada ukuran kualitatif gender male lebih besar dari female.

### Task 3: What is the the distribution of Age, Sex, BMI and Y variables?

Distribution dapat ditentukan dengan membuat histogram sehingga terlihat distribusi dari value sebuah data

```python
# Membuat histogram dengan ukuran bins 15
df['AGE'].hist(bins=15)
# Membuat judul plot histogram
plt.suptitle('AGE distribution of diabetes')
# Plot variabel X
plt.xlabel('AGE')
# Plot variabel Y
plt.ylabel('Count')
# Plot ditampilkan
plt.show()
```

Dari histogram 'AGE' dapat dilihat bahwa semua nilai berpusat di sekitar rata-rata tertentu. Distribusi ini tidak bersifat normal, tetapi sebagai histogram bimodal yang mana memiliki beberapa puncak tertinggi yang ketinggiannya tidak harus sama. Pada distribusi ini terlihat bahwa pasien diabetes didominasi oleh orang yang berkisaran umur 51-55.Dari distribusi ini, penyakit diabetes rentan diderita orang dewasa hingga lansia. Namun, penyakit ini juga dapat menyerang orang di usia muda di bawah 30 tahun. 

```python
# Membuat histogram dengan ukuran bins 15
df['SEX'].hist(bins=15)
# Membuat Judul dari plot histogram
plt.suptitle('SEX distribution of diabetes')
# plot variabel X
plt.xlabel('SEX')
# Plot variabel Y
plt.ylabel('Count')
# Plot ditampilkan
plt.show()
```

Dari 2 data dalam value 'SEX' sehingga datanya tidak terdistribusi normal dan terlihat bahwa distribusi data pasien diabeter didominasi oleh pria (male)dengan jumlah penderita diabetes yang paling banyak / mendominasi dibandingkan wanita (female).

```python
# Membuat histogram dengan ukuran bins 15
df['BMI'].hist(bins=15)
# Membuat Title atau Judul histogram
plt.suptitle('BMI distribution of diabetes')
# Plot variabel X
plt.xlabel('BMI')
# Plot variabel Y
plt.ylabel('Count')
# Plot ditampilkan
plt.show()
```

Distribusi value berat badan (BMI) bersifat right skew. Berdasarkan distribusinya, dapat diketahui bahwa Kemiringan data ke arah kanan (positif) dimana nilai mean lebih dari nilai modus. Dalam distribusi data ini terlihat bahwa penderita diabeter didominasi oleh orang yang memiliki BMI dalam kisaran 23-27. Dari hal ini jika dilihat dari referensi di internet sebagai berikut :

image.png

Dari kisaran BMI tersebut maka pasien diabetes didominasi oleh orang yang memiliki BMI normal dan overweight. Dominasi ini dipengaruhi oleh jumlah orang yang memiliki berat badan normal dan overweight lebih banyak dibandingkan yang memiliki berat dengan kondisi lainnya seperti obesitas sehingga jumlah pasiennya juga akan cenderung lebih banyak.


```python
# Membuat histogram dengan ukuran bins 15
df['Y'].hist(bins=15)
# Membuat title histogram
plt.suptitle('Y distribution of diabetes')
# Plot variabel X
plt.xlabel('Y')
# Plot variabel Y
plt.ylabel('Count')
# Plot ditampilkan
plt.show()
```

Distribusi ini dapat termasuk ke dalam right skew dengan memiliki beberapa puncak tertinggi yang ketinggiannya tidak sama. Pada kondisi ukuran kualitatif perkembangan penyakit selama satu tahun atau value 'Y' distribusinya didominasi oleh ukuran dalam kisaran 50-110.
