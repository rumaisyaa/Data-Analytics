# Basic Data Operation

You sold 20 pieces of electronics at the price of 12000. The customer gets a total discount of 10%.

Below are those variables:

1.   price=12000
2.   quantity_sold=20
3.   discount=10%


Using Python code, write the code the find the total money that the customer to pay. (Points : 15)

```python
#Code goes here
price = 12000
quantity_sold = 20
discount = 10/100
total = (price*quantity_sold)-(price*quantity_sold*discount)
total
```

**Penjelasan Jawaban Basic Data Operation**

Dari data variabel-variabel tersebut maka dibuatlah rumus untuk menentukan berapa harga yang harus customer bayarkan. Dari data tersebut, hal yang perlu dilakukan ialah menetukan keseluruhan harga dengan melakukan multiplication (price*quantity_sold) yang mana sebesar 240.000. Kemudian, perlu juga ditentukan harga keseluruhan yang telah didiskon sebesar 10% dengan rumus multiplication (price*quantity_sold*discount) sehingga didapatkan harga keseluruhan yang telah diberi diskon sebesar 24000. Dengan demikian, penentuan selisih dari harga keseluruhan dan harga diskon keseluruhan (substraction) didapat sebesar 216.000. Dari hasil itulah diketahui bahwa customer perlu membayarkan sejumlah uang sebesar 216.000.

**Penjelasan Jawaban Code No. 6 (5)**

Dalam mereplace value dalam list maka dapat dilakukan dengan membuat code berupa variabel dan key serta data value yang ingin diubah dalam list. Dari kasus ini, dibuatlah code customer['gender']=['male','female','female','male'] untuk mereplace gender 1 dengan male dan gender 0 untuk female. Dengan demikian value dalam list key 'gender' yang semulanya berupa '1','0','0','1' akan secara otomatis berubah valuenya di dalam dictionary tersebut ketika code dijalankan menjadi 'male','female','female','male'. Selain itu, jika replace datanya banyak maka dapat menggunakan for yang mana dapat mereplace secara otomatis data tersebut.

```python
#Tambahan code jawaban No 6(5) replace gender
#Tambahan code jika menggunakan cara for (if dan else)
#Code ini cocok digunakan jika data nya banyak
for x in customer['gender']:
  if x == 1:
    customer['gender'] = 'male'
  elif x == 0:
    customer['gender'] = 'female'

  
customer
```
