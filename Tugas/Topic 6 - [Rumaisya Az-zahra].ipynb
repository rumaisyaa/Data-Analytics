```python
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
                   
df2=pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                   'B': ['B0', 'B1', 'B2']})
```

```python
#import libraries here
#Import warning untuk mengabaikan pemberitahuan tidak penting
#Import pandas, dll

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Merunning dataframe 1 dari soal pertanyaan

df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
df1
```

```python
#Merunning dataframe 2 dari soal pertanyaan

df2=pd.DataFrame({'key': ['K0', 'K1', 'K2'],'B': ['B0', 'B1', 'B2']})
df2
```

### Create a dataframe (df3) which consists both dataframe inner-joined by column "Key" (5 pts)

```python
#code goes here
#Dataframe 3 dari inner join d1 dan d2
#Dilakukan merge dari relasi (hubungan) antara key

df3 = pd.merge(df1,df2,how="inner")
df3
```

### Create a dataframe (df4) which consists of the result of df1 and df2 left-joined by column "Key" (10pts)

```python
#code goes here
#Dataframe 4 didapatkan dari left join d1 dan d2
#semua data pada tabel sebelah kiri akan ditampilkan

df4 = pd.merge(df1,df2,how="left")
df4
```

#### 1. Replace key "K2" into "K3" on df1
#### 2. Left join it with df2 
(15pts)

```python
#code goes here
#Mereplace kata kunci "K2" menjadi "K3"
#Key "K2" akan hilang dan berubah

df1['key'] = df1['key'].replace(['K2'], 'K3')
df1
```

```python
#Df diperoleh dari left join d1 (key "K2" yang sudah direplace) dan d2
#semua data pada tabel sebelah kiri akan ditampilkan

df = pd.merge(df1,df2,how="left")
df
```

# Dataframe Processing


Create a dictionary and convert it to a dataframe called "Customer_df" (15 pts)


name=['Anna','Jason','Cindy']

age=['22','23','21']

gender=['F','M','F']




```python
#code goes here
#Membuat dataframe dari dictionary
#Dibuat menggunakan Join with key

dict={
    "name":['Anna','Jason','Cindy'],
    "age":[22,23,21],
    "gender" :['F','M','F'] 
}

customer_df=pd.DataFrame(dict)
customer_df
```

### Delete the "age" column (5 pts)

```python
#code goes here
#Menghapus kolom age dengan mendrop

customer_df = customer_df.drop('age',axis=1)
customer_df.head()
```

### Rename the "name" column into "first name" (10 pts)

```python
#code goes here
#Mengganti nama kolom menjadi "First name"
#Menggunakan cara 1 (inplace = True)

customer_df.rename(columns ={
    'name':'first name'
}, inplace = True)
customer_df.head()
```

### Sort the dataframe by age, descending (10 pts)

```python
#code goes here
#Karena kolom age sudah dihapus diatas maka perlu di running ulang

dict={
    "first name":['Anna','Jason','Cindy'],
    "age":[22,23,21],
    "gender" :['F','M','F'] 
}

customer_df=pd.DataFrame(dict)
customer_df
```

```python
#Mengurutkan data dari usia terbesar ke terkecil (menurun)

customer_df_temp = customer_df.sort_values(['age'], ascending=False)
customer_df_temp.head()
```

## Using data aggregation in pandas , Answer the question below:

Show the total amount of passengers who survived and not survived (10pts)

```python
#code goes here
#Jumlah penumpang yang selamat dan tidak

df.groupby('Survived').agg({'Survived':['sum']})

#Survived = 1
#Not Survived = 0
#Jumlah total penumpang selamat
```

Show all male survivors from titanic incident (10 pts)

```python
#code goes here
#Melihat jumlah laki-laki (male) yang selamat di insiden titanic

df.groupby('Sex').agg({'Survived':['sum']})
```

Show the average fare paid based on gender (10 pts)

```python
#code goes here
#Menghitung rata-rata harga tiket berdasar gender

df.groupby('Sex').agg({'Fare':['mean']})
```

Show the total amount of passengers who survived and not survived based on gender (10pts)
